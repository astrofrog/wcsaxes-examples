Initialization
--------------

The object-oriented interface (similar for ``plt.figure()`` and the true OO API) uses a WCSAxes class which takes the same argument as Axes, but with an additional argument for the WCS object:

```python
import matplotlib.pyplot as plt
from astropy.wcs import WCS
from astropy.io import fits
from wcsaxes import WCSAxes

hdu = fits.open('msx.fits')[0]

fig = plt.figure(figsize=(6,6))
ax = WCSAxes(fig, [0.1, 0.1, 0.8, 0.8], wcs=WCS(hdu.header))
fig.add_axes(ax)
```

By default, the region shown is from pixel positions 0 to 1 in both dimensions, as for a normal matplotlib plot, but showing the image then sets the view limits to the correct values:

```python
ax.imshow(hdu.data, vmin=-1e-5, vmax=1e-4, origin='lower')
fig
```

If no WCS transformation is specified, the transformation will default to identity, meaning that the WCSAxes object should then act like a normal Axes object:

```python
fig = plt.figure(figsize=(8,8))
ax = WCSAxes(fig, [0.1, 0.1, 0.8, 0.8])
fig.add_axes(ax)
```

Once ``WCSAxes`` is in Astropy, ``WCS`` will be modified to allow:

    from astropy import wcs
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1, projection=wcs.WCS('image.fits'))
    
Coordinate systems
------------------

As a convention, and to avoid any assumptions, if methods are called as:

    ax.method(...)
    
then the method applies to pixel coordinates:

```python
fig = plt.figure(figsize=(6,6))
ax = WCSAxes(fig, [0.1, 0.1, 0.8, 0.8], wcs=WCS(hdu.header))
fig.add_axes(ax)
ax.imshow(hdu.data, vmin=-1e-5, vmax=1e-4, origin='lower')
ax.scatter([100,120,150,90],[130,100,120,80], edgecolor='yellow',lw=2,facecolor='none', s=40)
```

 However, in general it will be useful to be able to plot object in various world coordinate systems. Matplotlib already supports plotting with arbitrary transformations, so a method of ``WCSAxes`` is available to give a transformation that can be directly used by Matplotlib:

```python
ax.get_transform("galactic")
```

To plot in Galactic coordinates, one would then do:

```python
ax.scatter([-0.1, 0.12, 0.03], [-0.25, -0.16, 0.15],
           edgecolor='orange', lw=2, facecolor='none', s=50,
           transform=ax.get_transform('galactic'))
fig
```

Contours
--------

The ``get_transform`` method can also take the WCS of another image and allow data from this image to be overplotted, for example as contours:

```python
# Read in 2MASS image in Equatorial coordinates
hdu = fits.open('2MASS_k.fits')[0]

# Read in MSX image in Galactic coordinates
hdu_msx = fits.open('msx.fits')[0]
wcs_msx = WCS(hdu_msx.header)

# Make colorscale figure
fig = plt.figure(figsize=(6,6))
ax = WCSAxes(fig, [0.1, 0.1, 0.8, 0.8], wcs=WCS(hdu.header))
fig.add_axes(ax)
ax.imshow(hdu.data, vmin=-100, vmax=3000, origin='lower')

# Overplot contour
ax.contour(hdu_msx.data, transform=ax.get_transform(wcs_msx), colors='orange', levels=[2.5e-5,5e-5,1.e-4])

ax.set_xlim(0., 720.)
ax.set_ylim(0., 720.)
```

```python
fig = plt.figure(figsize=(6,6))
ax = WCSAxes(fig, [0.1, 0.1, 0.8, 0.8], wcs=WCS(hdu.header))
fig.add_axes(ax)
ax.imshow(hdu.data, vmin=-100, vmax=3000, origin='lower')

# Overplot contour
ax.contourf(hdu_msx.data, transform=ax.get_transform(wcs_msx), cmap=plt.cm.gist_heat, levels=[2.5e-5,5e-5,1.e-4], alpha=0.5)

ax.set_xlim(0., 720.)
ax.set_ylim(0., 720.)
```

Ticks and tick label properties
-------------------------------

While for many images, the coordinate axes are aligned with the pixel axes, this is not always the case, especially in coordinate systems with high curvature, where the coupling between x- and y-axis to actual coordinates become less well-defined. Therefore rather than referring to x and y ticks (as in APLpy), we use a ``coords`` attribute to ``WCSAxes``, which allows very flexible customization of plots.

```python
hdu = fits.open('msx.fits')[0]
fig = plt.figure(figsize=(6,6))
ax = WCSAxes(fig, [0.1, 0.1, 0.8, 0.8], wcs=WCS(hdu.header))
fig.add_axes(ax)
ax.imshow(hdu.data, vmin=-1e-5, vmax=1e-4, origin='lower')
```

To access the coordinates, one can either use integers to indicate the order in the WCS:

```python
ax.coords[0]
```

or the name of the coordinate in the WCS:

```python
ax.coords['GLON']
```

### Tick label format

Using the above coordinate objects, we can now customize properties of the ticks and labels for that coordinate, starting with the label format:

```python
ax.coords['glon'].set_major_formatter('dd:mm')
ax.coords['glat'].set_major_formatter('dd:mm:ss.ssss')
ax.set_xlim(0,500)  # needed for now otherwise labels don't get refreshed
ax.set_ylim(0,500)
fig
```
